```python
import numpy as np
import pandas as pd
```

```python
hawks_df = pd.read_csv('./hawks.csv')
hawks_df.head()
```

```python
shape = hawks_df.shape 
shape
```

```python
lst_missing_ratio = []

for i in range(0,shape[1]):    
    column = hawks_df[hawks_df.columns[i]]
    miss_1 = (column == ' ').sum()
    miss_2 = (column.isna()).sum()
    miss_ratio = ((miss_1 + miss_2) / shape[0] * 100).round(1)
    print(miss_ratio,hawks_df.columns[i])    
    lst_missing_ratio.append(miss_ratio)
```

- Đối với những thuộc tính bị thiếu trên 50% thì xoá luôn thuộc tính đi  
    - ReleaseTime: thiếu 92.8%
    - Sex: thiếu 63.4%
    - Tarsus: thiếu 91.7%
    - WingPitFat: thiếu 91.5%
    
- Các giá trị thiếu của thuộc tính numeric sẽ được điền đầy bằng phương pháp mean
    - CaptureTime
    - Wing 
    - Weight
    - Culmen 
    - Hallux 
    - StandardTail
    - KeelFat 
    - Crop
- Các giá trị thiếu của thuộc tính nominal được điền đầy bằng phương pháp mode:
    - CaptureTime 
    
- Ta thấy thuộc tính BandNumber là thuộc tính index vì tất cả các dòng dữ liệu đều khác nhau. Do đó, ta xoá luôn cột này

```python
# xoá cột
hawks_df.drop(columns=['ReleaseTime', 'Sex', 'Tarsus', 'WingPitFat', 'BandNumber'], inplace=True)
hawks_df.head()
```

```python
# Đối với các cột numeric: thay missing values bằng mean

num_col = ['Wing' ,'Weight','Culmen' ,'Hallux' ,'StandardTail','KeelFat' ,'Crop']

for item in num_col:
    hawks_df[item] = hawks_df[item].replace(' ', np.nan, regex=True)
    hawks_df[item] = hawks_df[item].fillna(hawks_df[item].mean())

hawks_df.head()
```

```python
# Đối với các cột norminal: thay missing values bằng mode
hawks_df['CaptureTime'] = hawks_df['CaptureTime'].replace(' ', np.nan, regex=True)
hawks_df['CaptureTime'] = hawks_df['CaptureTime'].fillna((hawks_df['CaptureTime'].mode()[0]))
hawks_df.head()
```

```python
# lưu file

# set nominal attribute as class attribute
hawks_df = hawks_df[['Month', 'Day', 'Year', 'CaptureTime', 'Age', 'Wing','Weight',
                     'Culmen', 'Hallux', 'Tail', 'StandardTail', 'KeelFat', 'Crop', 'Species']] 

hawks_df.to_csv('Preprocess.csv', header=True, index=False)
```
